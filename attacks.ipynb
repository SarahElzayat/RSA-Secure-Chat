```python
from utils import *
import time
import matplotlib.pyplot as plt
```

```python
plaintext = alphabet_encryption('this is a sample plaintext')
times = []
sizes = []
```

```python
for keysize in range(4, 65, 2):
    public_key, private_key, n  = generate_key_pair(keysize)
    ciphertext = rsa_encryption(plaintext, public_key, n)
    n_prime_factors = []
    d = 2
    start = time.time()
    root_n = int(math.ceil(n**0.5))
    for i in range(root_n, 2, -1):
        if n%i == 0:
            n_prime_factors.append(i)
            n_prime_factors.append(n//i)
            break
            

    phi_n = (n_prime_factors[0] -1 ) *(n_prime_factors[1] -1 )
    # print(public_key, ph)
    calculated_private_key = pow(public_key, -1, phi_n)
    end = time.time()
    if calculated_private_key == private_key:
        times.append(end-start)
        sizes.append(keysize)
        print("keysize", keysize)
        print("time",end-start)
        print("n_prime_factors", n_prime_factors)
        print("actual private key", private_key)
        print("calculated private key", calculated_private_key)
    
```

```python
plt.plot(sizes, times, 'b')

plt.ylabel('Time (secs)')
plt.xlabel('Keysize (bits)')
plt.show()
```
