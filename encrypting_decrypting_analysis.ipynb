```python

import time
from utils import *
import matplotlib.pyplot as plt
```

```python
plaintext = alphabet_encryption("sample")
encryption_times =[]
decryption_times =[]
sizes = []
```

```python
for keysize in range(32, 2049, 2):
    sizes.append(keysize)
    pu, pr, n = generate_key_pair(keysize)
    
    encyption_start = time.time()
    ciphertext = rsa_encryption(plaintext, pu, n)
    encryption_end = time.time()
    encryption_times.append((encryption_end - encyption_start)*1000)
    
    decryption_start = time.time()
    decrypted_ciphertext =alphabet_decryption(rsa_decryption(ciphertext, pr, n))
    decryption_end = time.time()
    decryption_times.append((decryption_end - decryption_start)*1000)

    
    
```

```python
plt.plot(sizes, encryption_times, 'b')
plt.ylabel('Time (milliseconds)')
plt.xlabel('Keysize (bits)')
plt.title('Encryption time for different keysizes')
plt.show()

plt.plot(sizes, decryption_times, 'b')
plt.ylabel('Time (milliseconds)')
plt.xlabel('Keysize (bits)')
plt.title('Decryption time for different keysizes')
plt.show()

```
